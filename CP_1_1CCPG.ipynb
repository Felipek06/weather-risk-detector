```python
# 1. Criar um novo dataframe df1 sem as colunas Date Time.
df1 = df.drop(['Date', 'Time'], axis=1)
# 2. Visualizar df1.
df1.head()
```

```python
# 3. Converter todas as colunas em dados numéricos.
df1 = df1.apply(pd.to_numeric, errors='coerce')
# 4. Obter resultado do método .info()
df1.info()
```

```python
# 5. Analisar se existem dados faltantes e/ou zerados(isna e isnull). Faça a contagem desses registros.
df1.isna().sum()
```

```python
# 6. Se a quantidade desses registros for menor que 5% do total (2075259), excluir os registros
df1 = df1.dropna()
df1.isna().sum()
```

```python
# 7. Gerando uma amostra chamada df2 com 2% dos dados da coluna "global_active_power".
df2 = df1.sample(frac=0.02)
df2.head()
```
